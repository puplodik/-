```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
from google.colab import files
# Firstly we read our data
#  Upload file SPAM text message 20170820 - Data.csv in work space and save it in csv:
print('Загрузите файл SPAM text message 20170820 - Data.csv в рабочую область')
uploaded = files.upload()
data = pd.read_csv("SPAM text message 20170820 - Data.csv")
data.head()
#We must change category 1 or 0
data["Category"] = [1 if each == "spam" else 0 for each in data["Category"]]
data.head()
```

Выполнить предобработку текста, включая удаление знаков препинания, приведение к нижнему регистру и лемматизацию, чтобы подготовить текст для дальнейшей обработки и анализа

```python
#We choose 1 row.And we throw punctuation
import re
nlp_data = str(data.iloc[2,:])
nlp_data = re.sub("[^a-zA-Z]"," ",nlp_data)
#After return lower case
nlp_data = nlp_data.lower()
#we have two choice we can use split methot or tokenize
import nltk as nlp
import nltk
nltk.download('punkt')
nltk.download('wordnet')
nlp_data = nlp.word_tokenize(nlp_data)
#nlp_data = nlp_data.split() or we can do so
#we have to find word root
lemma = nlp.WordNetLemmatizer()
nlp_data = [lemma.lemmatize(word) for word in nlp_data]
#We join our data
nlp_data = " ".join(nlp_data)
```

Создать пустой список description_list, который будет использоваться для хранения обработанных текстовых описаний. Начать цикл по всем текстовым значениям, содержащимся в столбце «Message» набора данных data. Применить регулярное выражение, чтобы заменить все символы, кроме буквенных (a-z и A-Z), на пробелы. Использовать библиотеку NLTK для разделения строки на отдельные слова (токенизация).Создать объект лемматизатора из NLTK, который будет использоваться для преобразования слов к их базовым формам. Применить лемматизацию к каждому слову в списке, чтобы привести их к базовой форме. Лемматизированные слова объединить обратно в строку, разделяя их пробелами.

```python
import nltk as nlp
import re
description_list = []
for description in data["Message"]:
    description = re.sub("[^a-zA-Z]"," ",description)
    description = description.lower()   # buyuk harftan kucuk harfe cevirme
    description = nlp.word_tokenize(description)
    #description = [ word for word in description if not word in set(stopwords.words("english"))]
    lemma = nlp.WordNetLemmatizer()
    description = [ lemma.lemmatize(word) for word in description]
    description = " ".join(description)
    description_list.append(description) #we hide all word one section
```

Импортировать класс CountVectorizer из библиотеки scikit-learn, который используется для создания «мешка слов» (bag of words) из текстовых данных. Установить максимальное количество признаков (слов), которые будут использоваться в «мешке слов». Создать объект CountVectorizer с максимальным количеством признаков, установленным в max_features. Использовать CountVectorizer для преобразования списка обработанных текстовых описаний (description_list) в матрицу. Разделяем данные на обучающий (x_train, y_train) и тестовый (x_test, y_test) наборы в пропорции 90% к 10%.

```python
#We make bag of word it is including number of all word's info
from sklearn.feature_extraction.text import CountVectorizer
max_features = 3000 #We use the most common word
count_vectorizer = CountVectorizer(max_features = max_features, stop_words = "english")
sparce_matrix = count_vectorizer.fit_transform(description_list).toarray()
print("the most using {} words: {}".format(max_features,count_vectorizer.get_feature_names_out()))
#We separate our data is train and test
y = data.iloc[:,0].values   # male or female classes
x = sparce_matrix
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.1, random_state = 42)
```

Обучить модель GaussianNB на обучающем наборе данных x_train и соответствующих метках классов y_train. Вычислить и вывести точность (accuracy) модели GaussianNB на тестовом наборе данных x_test и метках классов y_test. Импортировать логистическую регрессию (Logistic Regression) из scikit-learn. Создать объект модели логистической регрессии, устанавливая максимальное количество итераций max_iter равным 200. Обучить модель логистической регрессии на обучающем наборе данных x_train и метках классов y_train. Вычислить и вывести точность модели логистической регрессии на тестовом наборе данных x_test и метках классов y_test. Создать объект модели K-Nearest Neighbors, устанавливая количество соседей n_neighbors равным 3. Обучить модель KNearest Neighbors. Вычислить и вывести точность модели.

```python
#We make model for predict
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(x_train,y_train)
print("the accuracy of our model: {}".format(nb.score(x_test,y_test)))
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(max_iter = 200)
lr.fit(x_train,y_train)
print("our accuracy is: {}".format(lr.score(x_test,y_test)))
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(x_train,y_train)
#print('Prediction: {}'.format(prediction))
print('With KNN (K=3) accuracy is: ',knn.score(x_test,y_test))
x_test = x_test.reshape(558,3000,1)
x_train = x_train.reshape(5014,3000,1)
```
